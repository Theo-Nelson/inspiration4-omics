```python
import pandas as pd, numpy as np, seaborn as sns
import re
from matplotlib.pyplot import subplots, style, rc
from collections import OrderedDict
from matplotlib.colors import PowerNorm, Normalize, LinearSegmentedColormap, to_rgba
from matplotlib.patches import Rectangle
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.cm import ScalarMappable

style.use(["seaborn-poster", "seaborn-whitegrid"])
rc("axes", linewidth=1, edgecolor="black")
%matplotlib inline
```

```python
dfnames2profiles = OrderedDict([
    ("postVpre1", "I4-FP1"),
    ("postVpre2", "I4-FP2"),
    ("postVpre3", "I4-FP3"),
    ("recVpre1",  "I4-LP1"),
    ("recVpre2",  "I4-LP2"),
    ("recVpre3",  "I4-LP4"),
    ("recVpre4",  "I4-LP5"),
    ("recVpre5",  "I4-LP6"),
    ("postVrec1", "I4-RP1"),
    ("postVrec2", "I4-RP2"),
    ("postVrec3", "I4-RP4"),
    ("postVrec4", "I4-RP5"),
    ("postVrec5", "I4-RP6"),
])
dfnames = list(dfnames2profiles)
FC_BREAKPOINT = 8
DISP = ["L-92", "L-44", "L-3", "R+1", "R+45", "R+82", "R+194"]
DISC = ["#68BBDA", "#377EB8", "#3E4D8A", "#E41A1C", "#D9742C", "#ECB03B", "#825F1D"]
```

```python
def zlog(df, vizgenes, normlib=True, divstd=True, pseudo=.01):
    if normlib:
        t = df.div(df.mean(axis=0))
    else:
        t = df.copy()
    t = np.log10(t + pseudo)
    if divstd:
        t = t.sub(t.mean(axis=1), axis=0).div(t.std(axis=1), axis=0).reindex(vizgenes)
    else:
        t = t.sub(t.mean(axis=1), axis=0).reindex(vizgenes)
    return t[sorted(t.columns, key=lambda x:x[1])].copy()
```

```python
def squishax(ax, factor=2, scalef=None):
    x0, y0, x1, y1 = ax.get_position().bounds
    adjust = (x1 - x0) * (factor-1) / factor
    ax.set_position([x0+adjust, y0, x1-adjust, y1])
    if scalef:
        ax.set_xscale("function", functions=(scalef, lambda _:_))
```

```python
TOFS = pd.read_csv("python/transcript_offsets.tsv.xz", sep="\t")
m6as = []

for name in dfnames:
    m6a_df = pd.read_csv(f"m6anet/methylKit/{name}-RTL.csv")
    if name.startswith("postVrec"):
        print("Fixing meth.diff direction in", name)
        m6a_df["meth.diff"] = -m6a_df["meth.diff"]
    m6a_df["ENST"] = m6a_df["chr"].apply(lambda s: s.split("|")[0])
    m6a_df["ENSG"] = m6a_df["chr"].apply(lambda s: s.split("|")[1])
    m6a_df = m6a_df.drop(columns=["chr", "end"]).rename(columns={"start": "ENST_position"})
    m6a_df = m6a_df.iloc[:,[6,5,0,1,2,3,4]].copy()
    m6a_fx = pd.merge(m6a_df.iloc[:,1:], TOFS, how="left")
    m6a_fx["ENSG_position"] = m6a_fx["ENST_position"] + m6a_fx["offset"]
    m6a_fx = m6a_fx.iloc[:,[0,1,2,6,8,3,4,5]].copy()
    m6a_fx.columns = pd.MultiIndex.from_tuples([
        ("transcript", "name"),
        ("transcript", "position"),
        ("transcript", "strand"),
        ("gene", "name"),
        ("gene", "position"),
        (name, "pvalue"),
        (name, "qvalue"),
        (name, "meth.diff"),
    ])
    m6a_fx = m6a_fx.set_index("transcript", drop=True)
    m6a_fx.index = m6a_fx.index.map(lambda ix: "|".join(map(str, ix)))
    m6as.append(m6a_fx)

_m6a = pd.concat([df.drop(columns="gene") for df in m6as], axis=1)

_transcriptlevel = _m6a.iloc[:,[]].reset_index()["transcript"].apply(lambda s: pd.Series(s.split("|")))
_transcriptlevel.columns = pd.MultiIndex.from_tuples([("transcript", s) for s in ("name", "position", "strand")])
_transcriptlevel.index = _m6a.index

_genelevel = pd.concat([df[["gene"]] for df in m6as], axis=0).reset_index().drop_duplicates().set_index("transcript")

mkit = pd.concat([_transcriptlevel, _genelevel, _m6a], axis=1)
mkit = mkit.loc[(mkit.iloc[:,list(range(6, mkit.shape[1]+1, 3))]<.01).any(axis=1)].copy()
```

```python
assert (mkit.iloc[:,2]=="+").all()

m6a_counts = pd.read_csv("m6anet/m6anet/m6anet-concat.csv.gz")
m6a_counts["ENST"] = m6a_counts["transcript_id"].apply(lambda s: s.split("|")[0])
m6a_counts = m6a_counts[m6a_counts["timepoint"]!="2021-11-pilot"].copy()
m6a_counts["locator"] = m6a_counts.apply(lambda r: "|".join([r["ENST"], str(r["transcript_position"]), "+"]), axis=1)
m6a_counts = m6a_counts.pivot(index="locator", columns=("subject", "timepoint"), values="probability_modified").copy()
m6a_counts.index.name = None
m6a_counts.columns.names = (None, None)
m6a_counts = m6a_counts[sorted(m6a_counts.columns, key=lambda x:x[1])].copy()

_m6a_ext = pd.concat([m6a_counts, mkit.reindex(m6a_counts.index).iloc[:, 5:]], axis=1).copy()

if True:
    t = _m6a_ext.copy()
    for subj in "C001", "C002", "C003", "C004":
        t[subj] = t[subj].fillna(0)
    t.iloc[:,range(28, t.shape[1], 3)] = t.iloc[:,range(28, t.shape[1], 3)].fillna(1)
    t.iloc[:,range(29, t.shape[1], 3)] = t.iloc[:,range(29, t.shape[1], 3)].fillna(1)
    t.columns = pd.MultiIndex.from_tuples([
        (b, a) if a.startswith("C00") else (a, b) for a, b in list(t.columns)
    ])
    def infer_md(compname, controls, targets):
        t[(compname, "meth.diff")] = (
            t[(compname, "meth.diff")]
            .fillna(t[targets].mean(axis=1) / t[controls].mean(axis=1))
            .apply(lambda v: float("nan") if abs(v) == float("inf") else v)
        )
    infer_md("postVpre1", ["2021-06", "2021-08", "2021-09-1pre"], ["2021-09-2post"])
    infer_md("postVpre2", ["2021-06", "2021-08", "2021-09-1pre"], ["2021-09-2post", "2021-11", "2021-12"])
    infer_md("postVpre3", ["2021-06", "2021-08", "2021-09-1pre"], ["2021-09-2post", "2021-11", "2021-12", "2022-03"])
    infer_md("recVpre1", ["2021-06", "2021-08", "2021-09-1pre"], ["2021-11"])
    infer_md("recVpre2", ["2021-06", "2021-08", "2021-09-1pre"], ["2021-12"])
    infer_md("recVpre3", ["2021-06", "2021-08", "2021-09-1pre"], ["2022-03"])
    infer_md("recVpre4", ["2021-06", "2021-08", "2021-09-1pre"], ["2021-12", "2022-03"])
    infer_md("recVpre5", ["2021-06", "2021-08", "2021-09-1pre"], ["2021-11", "2021-12", "2022-03"])
    infer_md("postVrec1", ["2021-11"], ["2021-09-2post"])
    infer_md("postVrec2", ["2021-12"], ["2021-09-2post"])
    infer_md("postVrec3", ["2022-03"], ["2021-09-2post"])
    infer_md("postVrec4", ["2021-12", "2022-03"], ["2021-09-2post"])
    infer_md("postVrec5", ["2021-11", "2021-12", "2022-03"], ["2021-09-2post"])
    t.columns = _m6a_ext.columns
    _m6a_ext.iloc[:,28:] = t.iloc[:,28:].copy() # _m6a_ext = t.copy()

_transcriptlevel = _m6a_ext.iloc[:,[]].reset_index()["index"].apply(lambda s: pd.Series(s.split("|")))
_transcriptlevel.columns = ["ENST", "ENST_position", "strand"]
_transcriptlevel["ENST_position"] = _transcriptlevel["ENST_position"].astype(int)
_transcriptlevel.index = _m6a_ext.index

_locator = pd.merge(_transcriptlevel, TOFS, how="left")
_locator["ENSG_position"] = _locator["ENST_position"] + _locator["offset"]
_locator.index = _locator.apply(lambda r: "|".join([r["ENST"], str(r["ENST_position"]), r["strand"]]), axis=1)
_locator = _locator.iloc[:,[0,1,2,3,5]].copy()
_locator.columns = pd.MultiIndex.from_tuples([
    *[("transcript", s) for s in ("name", "position", "strand")],
    *[("gene", s) for s in ("name", "position")],
])

m6a = pd.concat([_locator, _m6a_ext], axis=1).copy()
```

```python
mkit_filt = mkit.loc[(mkit.iloc[:,range(7, mkit.shape[1]+1, 3)]>10).any(axis=1)].copy()
viz_m6a = m6a.reindex(mkit_filt.index).iloc[:,5:33].copy()
viz_m6z = zlog(viz_m6a, mkit_filt.index, normlib=False, pseudo=.01)#.fillna(0)
```

```python
SUP, SDN = "▴", "▾"

meth_diff = m6a.reindex(viz_m6z.index).iloc[:,range(35, m6a.shape[1]+1, 3)].fillna(0).copy()
meth_diff.columns = meth_diff.columns.get_level_values(0)

meth_diff_q = m6a.reindex(viz_m6z.index).iloc[:,range(34, m6a.shape[1]+1, 3)].copy()
meth_diff_q.columns = meth_diff_q.columns.get_level_values(0)
meth_diff_direction = pd.DataFrame(index=meth_diff_q.index, columns=meth_diff_q.columns, data="")

for ix in meth_diff.index:
    for co in meth_diff.columns:
        if meth_diff_q.loc[ix, co] <.01:
            if meth_diff.loc[ix, co] > 10:
                meth_diff_direction.loc[ix, co] = SUP
            elif meth_diff.loc[ix, co] < -10:
                meth_diff_direction.loc[ix, co] = SDN
```

```python
m6aCmap = LinearSegmentedColormap.from_list(
    "m6aCmap", ((.0, "#A1DD8822"), (.2, "#A1DD8830"), (.4, "#A1DD8850"), (.8, "#BC567E"), (1., "#991155")),
)
m6aCmap
```

```python
DEBUG_PRINT_COORDS = False
DPI = 300

FHUG, FBIG, FMED, FSML = 28+2, 24+2, 20+2, 18
yfactor = 4.6
T1 = -6.65 * yfactor
H1 = 2.3 * yfactor
M1 = -5.4 * yfactor
T3 = -2.35 * yfactor
H3 = 2.35 * yfactor
M3 = -1.23 * yfactor
T2 = -4.35 * yfactor
H2 = 2 * yfactor
M2 = -3.3 * yfactor

# cluster backbone (meth_diff), set up colormap (m6aCmap):
vmax = viz_m6z.abs().max().max()
g = sns.clustermap(
    data=meth_diff, col_cluster=False, metric="correlation", method="ward",
    figsize=(25.4, 32), cmap=LinearSegmentedColormap.from_list("null", ((0, "white"), (1, "white"))),
    cbar_pos=None, dendrogram_ratio=.025,
)
squishax(g.ax_row_dendrogram, 1, lambda x: x**.5)

# paint over backbone with viz_m6z values:
sns.heatmap(
    data=viz_m6z.reindex(g.data2d.index),
    cmap=m6aCmap, norm=PowerNorm(2.3), vmin=-vmax, vmax=vmax,
    lw=1, cbar=False, ax=g.ax_heatmap,
)

# annotate counts/subjects sections:
hax = g.ax_heatmap
hax.set(xticks=[], xlabel=None)
hax.add_patch(Rectangle((0, T1), 12, H1, fc="#DAD9D9", ec="black", lw=1, clip_on=False))
hax.add_patch(Rectangle((12, T1), 4, H1, fc="#7E7C7D", ec="black", lw=1, clip_on=False))
hax.add_patch(Rectangle((16, T1), 12, H1, fc="#696667", ec="black", lw=1, clip_on=False))
hax.text(s="Pre-Flight", x=6, y=M1, ha="center", va="center", size=FHUG)
hax.text(s="Post", x=14, y=M1, ha="center", va="center", size=FHUG, color="w")
hax.text(s="Recovery", x=22, y=M1, ha="center", va="center", size=FHUG, color="w")
for i, kind in enumerate(["pre"]*3+["space"]+["rec"]*3):
    hax.add_patch(Rectangle((i*4, 0), 4, len(viz_m6z), fc="none", ec="black", lw=1, clip_on=False))
    hax.add_patch(Rectangle((i*4, T2), 4, H2, fc=DISC[i]+"DD", ec="black", lw=1, clip_on=False))
    hax.text(s=DISP[i], x=i*4+2, y=M2, color="white", ha="center", va="center", size=FHUG)
    for j in range(4):
        hax.add_patch(Rectangle((i*4+j, T3), 1, H3, fc="none", ec="gray", lw=1, clip_on=False))
        hax.text(s=f"C00{j+1}", x=i*4+j+.55, y=M3, ha="center", va="center", rotation=90, size=FMED)
    hax.add_patch(Rectangle((i*4, T2), 4, -T2+hax.get_ylim()[0], fc="none", ec="black", lw=1, clip_on=False))

# annotate underneath counts: custom colorbar
hax.text(
    s="$z$-$scores$ of log-transformed methylation probabilities", ha="left", va="bottom",
    x=0, y=hax.get_ylim()[0]+1.4*yfactor, size=FHUG,
)
# NOTE: the colorbar is meant to represent the colormap used above (with PowerNorm(2.3));
# however, here, if the same gamma (2.3) is used, it does not get centered around 0 at all,
# because matplotlib/seaborn want to show the colors stretched/squished according to the gamma.
# Instead, we reset this so that the zero is in the middle, by setting gamma to 1.
# IMPORTANT: this does **not** change the relationship of values to colors, but simply
# redistributes them uniformly in the colorbar!
scm = ScalarMappable(cmap=m6aCmap, norm=PowerNorm(1, vmin=-vmax, vmax=vmax))
for x, c in zip(np.linspace(0, 8.3, 200), np.linspace(-vmax, vmax, 200)):
    hax.plot(
        [x, x], [hax.get_ylim()[0]+1.6*yfactor, hax.get_ylim()[0]+2.5*yfactor],
        color=scm.to_rgba(c), clip_on=False,
    )
hax.text(s=f"{-vmax:.1f}", x=-.3, y=hax.get_ylim()[0]+2.9*yfactor, ha="left", va="top", size=FBIG)
hax.text(s="0", x=4.15, y=hax.get_ylim()[0]+2.9*yfactor, ha="center", va="top", size=FBIG)
hax.text(s=f"{vmax:.1f}", x=8.5, y=hax.get_ylim()[0]+2.9*yfactor, ha="right", va="top", size=FBIG)

# connect the two heatmaps:
hax.set(xticks=[], yticks=[], xlabel=None, ylabel=None)
for y in range(1, len(g.data2d)):
    hax.plot([28, 30], [y, y], clip_on=False, lw=1, color="lightgray")
hax.plot([28, 30], [0, 0], clip_on=False, lw=1, color="gray")
hax.plot([28, 30], [len(g.data2d)]*2, clip_on=False, lw=1, color="gray")

# print coordinates for debug:
if DEBUG_PRINT_COORDS:
    for y, ix in enumerate(g.data2d.index):
        hax.text(s=f"{y}", x=28.1, y=y+.62, ha="left", va="center", size=12)

# rearrange log2FC dataframes to align with clustermap; reverse log2FCs and directions for recVpost comparisons:
fx_meth_diff = meth_diff.reindex(g.data2d.index).copy()
fx_meth_diff.iloc[:,FC_BREAKPOINT:] = -fx_meth_diff.iloc[:,FC_BREAKPOINT:]
fx_meth_diff_direction = meth_diff_direction.reindex(g.data2d.index).copy()
fx_meth_diff_direction.iloc[:,FC_BREAKPOINT:] = fx_meth_diff_direction.iloc[:,FC_BREAKPOINT:].applymap(
    lambda arr: {SUP: SDN, SDN: SUP, "–": "–"}.get(arr, " "),
)

# WATCH OUT: switch positions of longitudinal and recovery
fx_meth_diff = fx_meth_diff.iloc[:,[0,1,2,8,9,10,11,12,3,4,5,6,7]]
fx_meth_diff_direction = fx_meth_diff_direction.iloc[:,[0,1,2,8,9,10,11,12,3,4,5,6,7]]

# add comparisons axis and plot log2FC heatmaps:
compax = make_axes_locatable(hax).append_axes("right", meth_diff.shape[1]*.566, pad=.5)
vmax = meth_diff.abs().max().max()
sns.heatmap(
    data=fx_meth_diff, cmap="PiYG_r", vmin=-vmax, vmax=vmax, cbar=False, lw=1,
    annot=fx_meth_diff_direction, fmt="s", annot_kws=dict(size=9, color="black", va="center"),
    ax=compax,
)
compax.set(xticks=[], yticks=[], xlabel=None, ylabel=None)
scm = ScalarMappable(cmap="PiYG_r", norm=Normalize(vmin=-vmax, vmax=vmax))

# annotate comparison names and groups:
for i, dfname in enumerate(fx_meth_diff_direction.columns):
    if i < 3:
        fc = "#949ECD"
    elif i < 8:
        fc = "#F8A88D"
    else:
        fc = "#94D2A0"
    compax.add_patch(Rectangle((i, T3), 1, H3, fc=fc+"77", ec="#FFFFFF88", lw=1, clip_on=False))
    compax.text(s=dfnames2profiles[dfname], x=i+.5, y=M3-.02, size=FMED-1, rotation=90, ha="center", va="center")
compax.add_patch(Rectangle((0, T3), fx_meth_diff.shape[1], -T3, fc="none", ec="black", lw=1, clip_on=False))
compax.add_patch(Rectangle(
    (0, T3), fx_meth_diff.iloc[:,:3].shape[1], -T3+fx_meth_diff.shape[0],
    fc="none", ec="black", lw=1, clip_on=False,
))
compax.add_patch(Rectangle(
    (3, T3), fx_meth_diff.iloc[:,3:FC_BREAKPOINT].shape[1], -T3+fx_meth_diff.shape[0],
    fc="none", ec="black", lw=1, clip_on=False,
))
compax.add_patch(Rectangle(
    (FC_BREAKPOINT, T3), fx_meth_diff.iloc[:,FC_BREAKPOINT:].shape[1], -T3+fx_meth_diff.shape[0],
    fc="none", ec="black", lw=1, clip_on=False,
))

compax.add_patch(Rectangle((0, T2), 3, T3-T2, fc="#949ECD", ec="black", lw=1, clip_on=False))
compax.text(s="Flight", x=1.5, y=M2, size=FHUG, ha="center", va="center")
compax.add_patch(Rectangle((3, T2), 5, T3-T2, fc="#F8A88D", ec="black", lw=1, clip_on=False))
compax.text(s="Recovery", x=5.5, y=M2, size=FHUG, ha="center", va="center")
compax.add_patch(Rectangle((8, T2), 5, T3-T2, fc="#94D2A0", ec="black", lw=1, clip_on=False))
compax.text(s="Longitudinal", x=10.5, y=M2, size=FHUG, ha="center", va="center")
compax.add_patch(Rectangle((0, T1), 13, H1, fc="#EACF8F", ec="black", lw=1, clip_on=False))
compax.text(s="Profile", x=6.5, y=M1, size=FHUG, ha="center", va="center")

# annotate underneath comparison heatmap: custom colorbar
compax.text(s="Δ $methylation$ %", x=0, y=compax.get_ylim()[0]+1.4*yfactor, ha="left", va="bottom", size=FHUG)
for x, c in zip(np.linspace(0, 6.5, 200), np.linspace(-vmax, vmax, 200)):
    compax.plot(
        [x, x], [compax.get_ylim()[0]+1.6*yfactor, compax.get_ylim()[0]+2.5*yfactor],
        color=scm.to_rgba(c), clip_on=False,
    )
compax.text(s=f"{-vmax:.1f}", x=-.25, y=compax.get_ylim()[0]+2.9*yfactor, ha="left", va="top", size=FBIG)
compax.text(s="0", x=3.25, y=compax.get_ylim()[0]+2.9*yfactor, ha="center", va="top", size=FBIG)
compax.text(s=f"{vmax:.1f}", x=6.6, y=compax.get_ylim()[0]+2.9*yfactor, ha="right", va="top", size=FBIG)
compax.set(xticks=[], yticks=[], xlabel=None, ylabel=None)

# annotate underneath comparison heatmap: explain triangles
xofs, yofs = 0, compax.get_ylim()[0] + 4.4*yfactor - 3
compax.add_patch(Rectangle((xofs+0, yofs+3), .9, 3.3, fc="lightgray", ec="white", lw=2, clip_on=False))
compax.text(s=SUP, ha="center", va="center", x=xofs+.3, y=yofs+4.6, size=FMED)
compax.text(s=SDN, ha="center", va="center", x=xofs+.6, y=yofs+4.6, size=FMED)
compax.text(s="significant", ha="left", va="center", x=xofs+1, y=yofs+4.7, size=FBIG)

yofs += 2.32
compax.add_patch(Rectangle((xofs+0, yofs+4.5), .9, 3.3, fc="lightgray", ec="white", lw=2, clip_on=False))
compax.text(s="below significance", ha="left", va="center", x=xofs+1, y=yofs+6.2, size=FBIG)

# oh, also add 'NA' under counts:
hax.add_patch(Rectangle(
    (9.2, hax.get_ylim()[0]+1.6*yfactor), .75, .9*yfactor,
    fc="white", ec="black", lw=1.5, clip_on=False,
))
hax.text(s=f"$NA$", x=9.2+.75/2, y=hax.get_ylim()[0]+2.9*yfactor, ha="center", va="top", size=FBIG)

# annotate gene sets:
genesettextkw = dict(rotation=90, ha="center", va="center", color="black", size=FHUG)
compax.add_patch(Rectangle((13.5, 0), .75, 56, fc="#DDD", ec="#DDD", lw=1, clip_on=False))
compax.text(s="$Set$ $M$-$I$", x=13.95, y=(0+56)/2, **genesettextkw)
compax.add_patch(Rectangle((13.5, 56), .75, 213-56, fc="#C2C2C2", ec="#C2C2C2", lw=1, clip_on=False))
compax.text(s="$Set$ $M$-$II$", x=13.95, y=(56+213)/2, **genesettextkw)
compax.add_patch(Rectangle((13.5, 213), .75, 305-213, fc="#A2A2A2", ec="#A2A2A2", lw=1, clip_on=False))
compax.text(s="$Set$ $M$-$III$", x=13.95, y=(213+305)/2, **genesettextkw)

g.fig.dpi = DPI
```
